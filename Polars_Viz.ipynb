```python
# import gapminder dataset
from gapminder import gapminder
# read the head
gapminder.head()
```

**import polars:**

```python
import polars as pl
```

```python
pl.show_versions()
```

**from pandas to polar:**

```python
# read dataset from pandas dataframe to polars dataframe
df_gapminder = pl.from_pandas(gapminder)
```

```python
# using seaborn
import seaborn as sns

ax = sns.scatterplot(
    df_gapminder,
     x="lifeExp",
    y="gdpPercap",
    hue="continent",
    style="continent"
);
ax.set(title="Life Expectancy vs GDP Per Capita")
sns.move_legend(ax, "upper left")
```
